# Red-band objective regression

Full-grid phase optimization against a historical red-detuned band objective. This numerical regression does not attribute the change to Raman physics. The demo path is problem, control, objective, solve, report.

```julia
import Pkg
project_root = isfile("Project.toml") ? pwd() : dirname(pwd())
Pkg.activate(project_root; io = devnull)
cd(project_root)

using Base64
using FiberLab

function show_report(report)
    println("Standard figures are embedded in the notebook below.")
    report
end
```

```julia
fiber = Fiber(preset = :SMF28, length_m = 2.0, power_w = 0.2)
grid = Grid(nt = 1024, time_window_ps = 12.0, policy = :exact)

problem = fiber_problem(fiber; grid = grid, raman_threshold_thz = -13.2)
control = phase_control(problem)
objective = raman_band_objective(problem)

dimension(control), summarize(problem)
```

```julia
result = solve(
    problem,
    control,
    objective;
    id = "raman_band_demo",
    max_iter = 4,
    step_size = 1e-2,
    maturity = :supported,
)

report = standard_report(
    problem,
    result;
    output_dir = joinpath("examples", "outputs", "01_red_band_objective"),
    tag = "raman_band_demo",
    n_z_samples = 72,
)

display_report(report)
metrics(result)
```

```julia
show_report(report)
```
