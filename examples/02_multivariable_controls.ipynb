# Multivariable controls

Phase, amplitude, and energy optimized together. Bounds live in the control definitions, not in ad hoc notebook glue.

```julia
import Pkg
project_root = isfile("Project.toml") ? pwd() : dirname(pwd())
Pkg.activate(project_root; io = devnull)
cd(project_root)

using Base64
using FiberLab

function show_report(report)
    println("Standard figures are embedded in the notebook below.")
    report
end
```

```julia
fiber = Fiber(preset = :SMF28, length_m = 2.0, power_w = 0.2)
grid = Grid(nt = 1024, time_window_ps = 12.0, policy = :exact)

problem = fiber_problem(fiber; grid = grid, raman_threshold_thz = -13.2)
basis = polynomial_basis(problem, 0:3)

control = controls(
    phase_control(problem; basis = basis, bounds = (-3.0, 3.0)),
    amplitude_control(problem; basis = basis, bounds = (0.8, 1.2)),
    energy_control(),
)
objective = raman_band_objective(problem)

dimension(control), summarize(problem)
```

```julia
result = solve(
    problem,
    control,
    objective;
    id = "multivariable_demo",
    max_iter = 6,
    step_size = 1e-2,
    maturity = :supported,
)

report = standard_report(
    problem,
    result;
    output_dir = joinpath("examples", "outputs", "02_multivariable_controls"),
    tag = "multivariable_demo",
    n_z_samples = 72,
)

display_report(report)
metrics(result)
```

```julia
show_report(report)
```
