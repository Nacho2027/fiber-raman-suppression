# Reduced-basis phase optimization

A low-dimensional phase parameterization uses the same solve and report calls as the full-grid phase case.

```julia
import Pkg
project_root = isfile("Project.toml") ? pwd() : dirname(pwd())
Pkg.activate(project_root; io = devnull)
cd(project_root)

using Base64
using FiberLab

function show_report(report)
    println("Standard figures are embedded in the notebook below.")
    report
end
```

```julia
fiber = Fiber(preset = :SMF28, length_m = 2.0, power_w = 0.2)
grid = Grid(nt = 1024, time_window_ps = 12.0, policy = :exact)

problem = fiber_problem(fiber; grid = grid, raman_threshold_thz = -13.2)
basis = fourier_basis(problem, 8)
control = phase_control(problem; basis = basis, bounds = (-4.0, 4.0))
objective = raman_band_objective(problem)

dimension(control), summarize(problem)
```

```julia
result = solve(
    problem,
    control,
    objective;
    id = "reduced_basis_demo",
    max_iter = 8,
    step_size = 1e-2,
)

report = standard_report(
    problem,
    result;
    output_dir = joinpath("examples", "outputs", "04_reduced_basis_phase"),
    tag = "reduced_basis_demo",
    n_z_samples = 72,
)

display_report(report)
metrics(result)
```

```julia
show_report(report)
```
